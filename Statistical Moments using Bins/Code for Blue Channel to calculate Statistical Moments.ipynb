```python
from PIL import Image
import numpy as np
from skimage import io
from math import sqrt
from numpy import mean
import csv
import cv2
import os

MEAN_B = [0 for i in range(8)]
STD_B = [0 for i in range(8)]
VAR_B = [0 for i in range(8)]
SKEW_B = [0 for i in range(8)]
KURTO_B = [0 for i in range(8)]
bins = [0 for i in range(8)]
bin_pixels_b = [[] for i in range(8)]
B_STD_BIN8 = [0 for i in range(8)]
B_VAR_BIN8 = [0 for i in range(8)]
B_SKEW_BIN8 = [0 for i in range(8)]
B_KURTO_BIN8 = [0 for i in range(8)]

def getimagedata(path):
    filename = os.listdir(path)
    n = len(filename)
    data = np.zeros([n,32])
    with open('image_bins_stats_b_p.csv', 'w', newline='') as f:
        hey = csv.writer(f)
        hey.writerow(['bmean_bins0','bmean_bins1','bmean_bins2','bmean_bins3','bmean_bins4','bmean_bins5','bmean_bins6','bmean_bins7',\
                      'bstd_bins0','bstd_bins1','bstd_bins2','bstd_bins3','bstd_bins4','bstd_bins5','bstd_bins6','bstd_bins7',\
                      'bskew_bins0','bskew_bins1','bskew_bins2','bskew_bins3','bskew_bins4','bskew_bins5','bskew_bins6','bskew_bins7',\
                      'bkurto_bins0','bkurto_bins1','bkurto_bins2','bkurto_bins3','bkurto_bins4','bkurto_bins5','bkurto_bins6','bkurto_bins7'])
        for i in range(n):
            img = io.imread(path +'\\'+ filename[i])
            a = np.array(img[:,:,0], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogr = num / den
        
            a = np.array(img[:,:,1], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogg = num / den
        
            a = np.array(img[:,:,2], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogb = num / den
            
            image1 = np.array(img[:,:,0], dtype = np.uint8)
            image2 = np.array(img[:,:,1], dtype = np.uint8)
            image3 = np.array(img[:,:,2], dtype = np.uint8)
            h, w = image1.shape
            rplane = np.copy(image1)
            for r in range(h):
                for c in range(w):
                    if rplane[r, c] <= cogr:
                        rplane[r, c] = 0
                    else:
                        rplane[r, c] = 1
                        
            h, w = image2.shape           
            gplane = np.copy(image2)
            for r in range(h):
                for c in range(w):
                    if gplane[r, c] <= cogg:
                        gplane[r, c] = 0
                    else:
                        gplane[r, c] = 1
            
            h, w = image3.shape
            bplane = np.copy(image3)
            for r in range(h):
                for c in range(w):
                    if bplane[r, c] <= cogb:
                        bplane[r, c] = 0
                    else:
                        bplane[r, c] = 1

            for i in range(h):
                for j in range(w):
                    r = rplane[i, j]
                    g = gplane[i, j]
                    b = bplane[i, j]

                    # image1 for rplane, image2 for gplane, image 3 for bplane
                    if r == 0 and g == 0 and b == 0:
                        bins[0] = bins[0] + 1
                        bin_pixels_b[0].append(image3[i, j])
                        MEAN_B[0] =  mean(bin_pixels_b[0])
                        VAR_B[0] = VAR_B[0] + abs(image3[i, j] - mean(bin_pixels_b[0])) ** 2
                        SKEW_B[0] = SKEW_B[0] + abs(image3[i, j] - mean(bin_pixels_b[0])) ** 3
                        KURTO_B[0] = KURTO_B[0] + abs(image3[i, j] - mean(bin_pixels_b[0])) ** 4

                    elif r == 0 and g == 0 and b == 1:
                        bins[1] = bins[1] + 1
                        bin_pixels_b[1].append(image3[i, j])
                        MEAN_B[1] =  mean(bin_pixels_b[1])
                        VAR_B[1] = VAR_B[1] + abs(image3[i, j] - mean(bin_pixels_b[1])) ** 2
                        SKEW_B[1] = SKEW_B[1] + abs(image3[i, j] - mean(bin_pixels_b[1])) ** 3
                        KURTO_B[1] = KURTO_B[1] + abs(image3[i, j] - mean(bin_pixels_b[1])) ** 4
                        
                    elif r == 0 and g == 1 and b == 0:
                        bins[2] = bins[2] + 1
                        bin_pixels_b[2].append(image3[i, j])
                        MEAN_B[2] =  mean(bin_pixels_b[2])
                        VAR_B[2] = VAR_B[2] + abs(image3[i, j] - mean(bin_pixels_b[2])) ** 2
                        SKEW_B[2] = SKEW_B[2] + abs(image3[i, j] - mean(bin_pixels_b[2])) ** 3
                        KURTO_B[2] = KURTO_B[2] + abs(image3[i, j] - mean(bin_pixels_b[2])) ** 4
                        
                    elif r == 0 and g == 1 and b == 1:
                        bins[3] = bins[3] + 1
                        bin_pixels_b[3].append(image3[i, j])
                        MEAN_B[3] =  mean(bin_pixels_b[3])
                        VAR_B[3] = VAR_B[3] + abs(image3[i, j] - mean(bin_pixels_b[3])) ** 2
                        SKEW_B[3] = SKEW_B[3] + abs(image3[i, j] - bin_pixels_b[3].mean()) ** 3
                        KURTO_B[3] = KURTO_B[3] + abs(image3[i, j] - bin_pixels_b[3].mean()) ** 4

                    elif r == 1 and g == 0 and b == 0:
                        bins[4] = bins[4] + 1
                        bin_pixels_b[4].append(image3[i, j])
                        MEAN_B[4] =  mean(bin_pixels_b[4])
                        VAR_B[4] = VAR_B[4] + abs(image3[i, j] - mean(bin_pixels_b[4])) ** 2
                        SKEW_B[4] = SKEW_B[4] + abs(image3[i, j] - mean(bin_pixels_b[4])) ** 3
                        KURTO_B[4] = KURTO_B[4] + abs(image3[i, j] - mean(bin_pixels_b[4])) ** 4

                    elif r == 1 and g == 0 and b == 1:
                        bins[5] = bins[5] + 1
                        bin_pixels_b[5].append(image3[i, j])
                        MEAN_B[5] =  mean(bin_pixels_b[5])
                        VAR_B[5] = VAR_B[5] + abs(image3[i, j] - mean(bin_pixels_b[5])) ** 2
                        SKEW_B[5] = SKEW_B[5] + abs(image3[i, j] - mean(bin_pixels_b[5])) ** 3
                        KURTO_B[5] = KURTO_B[5] + abs(image3[i, j] - mean(bin_pixels_b[5])) ** 4
                        
                    elif r == 1 and g == 1 and b == 0:
                        bins[6] = bins[6] + 1
                        bin_pixels_b[6].append(image3[i, j])
                        MEAN_B[6] =  mean(bin_pixels_b[6])
                        VAR_B[6] = VAR_B[6] + abs(image3[i, j] - mean(bin_pixels_b[6])) ** 2
                        SKEW_B[6] = SKEW_B[6] + abs(image3[i, j] - mean(bin_pixels_b[6])) ** 3
                        KURTO_B[6] = KURTO_B[6] + abs(image3[i, j] - mean(bin_pixels_b[6])) ** 4

                    elif r == 1 and g == 1 and b == 1:
                        bins[7] = bins[7] + 1
                        bin_pixels_b[7].append(image3[i, j])
                        MEAN_B[7] =  mean(bin_pixels_b[7])
                        VAR_B[7] = VAR_B[7] + abs(image3[i, j] - mean(bin_pixels_b[7])) ** 2
                        SKEW_B[7] = SKEW_B[7] + abs(image3[i, j] - mean(bin_pixels_b[7])) ** 3
                        KURTO_B[7] = KURTO_B[7] + abs(image3[i, j] - mean(bin_pixels_b[7])) ** 4
                        
            for i in range(8):
                if bins[i] == 0:
                    Bstd = 0
                    B_STD_BIN8[i] = Bstd
                    Bvar = 0
                    B_VAR_BIN8[i] = Bvar
                    Bskew = 0
                    B_SKEW_BIN8[i] = Bskew
                    Bkurto = 0
                    B_KURTO_BIN8[i] = Bkurto
                    
                else:
                    pixelcount = bins[i]
                    B_VAR_BIN8[i] = VAR_B[i]/pixelcount
                    Bstd = sqrt(VAR_B[i]/pixelcount)
                    B_STD_BIN8[i] = Bstd
                    Bskew = (SKEW_B[i]/pixelcount)**(1/3.0)
                    B_SKEW_BIN8[i] = Bskew
                    Bkurto = (KURTO_B[i]/pixelcount)**(1/4.0)
                    B_KURTO_BIN8[i] = Bkurto

            '''
            ['rmean_bins0','rmean_bins1','rmean_bins2','rmean_bins3','rmean_bins4','rmean_bins5','rmean_bins6','rmean_bins7',/
                      'rstd_bins0','rstd_bins1','rstd_bins2','rstd_bins3','rstd_bins4','rstd_bins5','rstd_bins6','rstd_bins7',/
                      'rskew_bins0','rskew_bins1','rskew_bins2','rskew_bins3','rskew_bins4','rskew_bins5','rskew_bins6','rskew_bins7',/
                      'rkurto_bins0','rkurto_bins1','rkurto_bins2','rkurto_bins3','rkurto_bins4','rkurto_bins5','rkurto_bins6','rkurto_bins7']
            '''
            hey.writerow([MEAN_B[0],MEAN_B[1],MEAN_B[2],MEAN_B[3],MEAN_B[4],MEAN_B[5],MEAN_B[6],MEAN_B[7],\
                          B_STD_BIN8[0],B_STD_BIN8[1],B_STD_BIN8[2],B_STD_BIN8[3],B_STD_BIN8[4],B_STD_BIN8[5],B_STD_BIN8[6],B_STD_BIN8[7],\
                          B_SKEW_BIN8[0],B_SKEW_BIN8[1],B_SKEW_BIN8[2],B_SKEW_BIN8[3],B_SKEW_BIN8[4],B_SKEW_BIN8[5],B_SKEW_BIN8[6],B_SKEW_BIN8[7],\
                         B_KURTO_BIN8[0],B_KURTO_BIN8[1],B_KURTO_BIN8[2],B_KURTO_BIN8[3],B_KURTO_BIN8[4],B_KURTO_BIN8[5],B_KURTO_BIN8[6],B_KURTO_BIN8[7]])
        return data

if __name__=='__main__':
    a = getimagedata(r"C:\Users\Admin\Desktop\Statistical Moments using Bins\sample")
```
