```python
from PIL import Image
import numpy as np
from skimage import io
from math import sqrt
from numpy import mean
import csv
import cv2
import os

MEAN_R = [0 for i in range(8)]
STD_R = [0 for i in range(8)]
VAR_R = [0 for i in range(8)]
SKEW_R = [0 for i in range(8)]
KURTO_R = [0 for i in range(8)]
bins = [0 for i in range(8)]
bin_pixels_r = [[] for i in range(8)]
R_STD_BIN8 = [0 for i in range(8)]
R_VAR_BIN8 = [0 for i in range(8)]
R_SKEW_BIN8 = [0 for i in range(8)]
R_KURTO_BIN8 = [0 for i in range(8)]

def getimagedata(path):
    filename = os.listdir(path)
    n = len(filename)
    data = np.zeros([n,32])
    with open('image_bins_stats_r_p.csv', 'w', newline='') as f:
        hey = csv.writer(f)
        hey.writerow(['rmean_bins0','rmean_bins1','rmean_bins2','rmean_bins3','rmean_bins4','rmean_bins5','rmean_bins6','rmean_bins7',\
                      'rstd_bins0','rstd_bins1','rstd_bins2','rstd_bins3','rstd_bins4','rstd_bins5','rstd_bins6','rstd_bins7',\
                      'rskew_bins0','rskew_bins1','rskew_bins2','rskew_bins3','rskew_bins4','rskew_bins5','rskew_bins6','rskew_bins7',\
                      'rkurto_bins0','rkurto_bins1','rkurto_bins2','rkurto_bins3','rkurto_bins4','rkurto_bins5','rkurto_bins6','rkurto_bins7'])
        for i in range(n):
            img = io.imread(path +'\\'+ filename[i])
            a = np.array(img[:,:,0], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogr = num / den
        
            a = np.array(img[:,:,1], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogg = num / den
        
            a = np.array(img[:,:,2], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogb = num / den
            
            image1 = np.array(img[:,:,0], dtype = np.uint8)
            image2 = np.array(img[:,:,1], dtype = np.uint8)
            image3 = np.array(img[:,:,2], dtype = np.uint8)
            h, w = image1.shape
            rplane = np.copy(image1)
            for r in range(h):
                for c in range(w):
                    if rplane[r, c] <= cogr:
                        rplane[r, c] = 0
                    else:
                        rplane[r, c] = 1
                        
            h, w = image2.shape           
            gplane = np.copy(image2)
            for r in range(h):
                for c in range(w):
                    if gplane[r, c] <= cogg:
                        gplane[r, c] = 0
                    else:
                        gplane[r, c] = 1
            
            h, w = image3.shape
            bplane = np.copy(image3)
            for r in range(h):
                for c in range(w):
                    if bplane[r, c] <= cogb:
                        bplane[r, c] = 0
                    else:
                        bplane[r, c] = 1

            for i in range(h):
                for j in range(w):
                    r = rplane[i, j]
                    g = gplane[i, j]
                    b = bplane[i, j]

                    # image1 for rplane, image2 for gplane, image 3 for bplane
                    if r == 0 and g == 0 and b == 0:
                        bins[0] = bins[0] + 1
                        bin_pixels_r[0].append(image1[i, j])
                        MEAN_R[0] =  mean(bin_pixels_r[0])
                        VAR_R[0] = VAR_R[0] + abs(image1[i, j] - mean(bin_pixels_r[0])) ** 2
                        SKEW_R[0] = SKEW_R[0] + abs(image1[i, j] - mean(bin_pixels_r[0])) ** 3
                        KURTO_R[0] = KURTO_R[0] + abs(image1[i, j] - mean(bin_pixels_r[0])) ** 4
                        
                    elif r == 0 and g == 0 and b == 1:
                        bins[1] = bins[1] + 1
                        bin_pixels_r[1].append(image1[i, j])
                        MEAN_R[1] =  mean(bin_pixels_r[1])
                        VAR_R[1] = VAR_R[1] + abs(image1[i, j] - mean(bin_pixels_r[1])) ** 2
                        SKEW_R[1] = SKEW_R[1] + abs(image1[i, j] - mean(bin_pixels_r[1])) ** 3
                        KURTO_R[1] = KURTO_R[1] + abs(image1[i, j] - mean(bin_pixels_r[1])) ** 4
                
                    elif r == 0 and g == 1 and b == 0:
                        bins[2] = bins[2] + 1
                        bin_pixels_r[2].append(image1[i, j])
                        MEAN_R[2] =  mean(bin_pixels_r[2])
                        VAR_R[2] = VAR_R[2] + abs(image1[i, j] - mean(bin_pixels_r[2])) ** 2
                        SKEW_R[2] = SKEW_R[2] + abs(image1[i, j] - mean(bin_pixels_r[2])) ** 3
                        KURTO_R[2] = KURTO_R[2] + abs(image1[i, j] - mean(bin_pixels_r[2])) ** 4


                    elif r == 0 and g == 1 and b == 1:
                        bins[3] = bins[3] + 1
                        bin_pixels_r[3].append(image1[i, j])
                        MEAN_R[3] =  mean(bin_pixels_r[3])
                        VAR_R[3] = VAR_R[3] + abs(image1[i, j] - mean(bin_pixels_r[3])) ** 2
                        SKEW_R[3] = SKEW_R[3] + abs(image1[i, j] - bin_pixels_r[3].mean()) ** 3
                        KURTO_R[3] = KURTO_R[3] + abs(image1[i, j] - bin_pixels_r[3].mean()) ** 4

                    elif r == 1 and g == 0 and b == 0:
                        bins[4] = bins[4] + 1
                        bin_pixels_r[4].append(image1[i, j])
                        MEAN_R[4] =  mean(bin_pixels_r[4])
                        VAR_R[4] = VAR_R[4] + abs(image1[i, j] - mean(bin_pixels_r[4])) ** 2
                        SKEW_R[4] = SKEW_R[4] + abs(image1[i, j] - mean(bin_pixels_r[4])) ** 3
                        KURTO_R[4] = KURTO_R[4] + abs(image1[i, j] - mean(bin_pixels_r[4])) ** 4
                        
                    elif r == 1 and g == 0 and b == 1:
                        bins[5] = bins[5] + 1
                        bin_pixels_r[5].append(image1[i, j])
                        MEAN_R[5] =  mean(bin_pixels_r[5])
                        VAR_R[5] = VAR_R[5] + abs(image1[i, j] - mean(bin_pixels_r[5])) ** 2
                        SKEW_R[5] = SKEW_R[5] + abs(image1[i, j] - mean(bin_pixels_r[5])) ** 3
                        KURTO_R[5] = KURTO_R[5] + abs(image1[i, j] - mean(bin_pixels_r[5])) ** 4

                    elif r == 1 and g == 1 and b == 0:
                        bins[6] = bins[6] + 1
                        bin_pixels_r[6].append(image1[i, j])
                        MEAN_R[6] =  mean(bin_pixels_r[6])
                        VAR_R[6] = VAR_R[6] + abs(image1[i, j] - mean(bin_pixels_r[6])) ** 2
                        SKEW_R[6] = SKEW_R[6] + abs(image1[i, j] - mean(bin_pixels_r[6])) ** 3
                        KURTO_R[6] = KURTO_R[6] + abs(image1[i, j] - mean(bin_pixels_r[6])) ** 4
                        
                    elif r == 1 and g == 1 and b == 1:
                        bins[7] = bins[7] + 1
                        bin_pixels_r[7].append(image1[i, j])
                        MEAN_R[7] =  mean(bin_pixels_r[7])
                        VAR_R[7] = VAR_R[7] + abs(image1[i, j] - mean(bin_pixels_r[7])) ** 2
                        SKEW_R[7] = SKEW_R[7] + abs(image1[i, j] - mean(bin_pixels_r[7])) ** 3
                        KURTO_R[7] = KURTO_R[7] + abs(image1[i, j] - mean(bin_pixels_r[7])) ** 4
            for i in range(8):
                if bins[i] == 0:
                    Rstd = 0
                    R_STD_BIN8[i] = Rstd
                    Rvar = 0
                    R_VAR_BIN8[i] = Rvar
                    Rskew = 0
                    R_SKEW_BIN8[i] = Rskew
                    Rkurto = 0
                    R_KURTO_BIN8[i] = Rkurto
                else:
                    pixelcount = bins[i]
                    R_VAR_BIN8[i] = VAR_R[i]/pixelcount
                    Rstd = sqrt(VAR_R[i]/pixelcount)
                    R_STD_BIN8[i] = Rstd
                    Rskew = (SKEW_R[i]/pixelcount)**(1/3.0)
                    R_SKEW_BIN8[i] = Rskew
                    Rkurto = (KURTO_R[i]/pixelcount)**(1/4.0)
                    R_KURTO_BIN8[i] = Rkurto
            '''
            ['rmean_bins0','rmean_bins1','rmean_bins2','rmean_bins3','rmean_bins4','rmean_bins5','rmean_bins6','rmean_bins7',/
                      'rstd_bins0','rstd_bins1','rstd_bins2','rstd_bins3','rstd_bins4','rstd_bins5','rstd_bins6','rstd_bins7',/
                      'rskew_bins0','rskew_bins1','rskew_bins2','rskew_bins3','rskew_bins4','rskew_bins5','rskew_bins6','rskew_bins7',/
                      'rkurto_bins0','rkurto_bins1','rkurto_bins2','rkurto_bins3','rkurto_bins4','rkurto_bins5','rkurto_bins6','rkurto_bins7']
            '''
            hey.writerow([MEAN_R[0],MEAN_R[1],MEAN_R[2],MEAN_R[3],MEAN_R[4],MEAN_R[5],MEAN_R[6],MEAN_R[7],\
                          R_STD_BIN8[0],R_STD_BIN8[1],R_STD_BIN8[2],R_STD_BIN8[3],R_STD_BIN8[4],R_STD_BIN8[5],R_STD_BIN8[6],R_STD_BIN8[7],\
                          R_SKEW_BIN8[0],R_SKEW_BIN8[1],R_SKEW_BIN8[2],R_SKEW_BIN8[3],R_SKEW_BIN8[4],R_SKEW_BIN8[5],R_SKEW_BIN8[6],R_SKEW_BIN8[7],\
                         R_KURTO_BIN8[0],R_KURTO_BIN8[1],R_KURTO_BIN8[2],R_KURTO_BIN8[3],R_KURTO_BIN8[4],R_KURTO_BIN8[5],R_KURTO_BIN8[6],R_KURTO_BIN8[7]])
            f.flush()
        return data

if __name__=='__main__':
    a = getimagedata(r"C:\Users\Admin\Desktop\Statistical Moments using Bins\sample")
```
