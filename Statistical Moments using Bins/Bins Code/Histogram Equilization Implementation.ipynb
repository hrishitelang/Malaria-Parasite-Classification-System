```python
from os import listdir, path
from natsort import natsorted
from skimage import io
import numpy as np
import csv
def getbins(cell_images_path):
    cell_image_files = listdir(cell_images_path)
    cell_image_files = natsorted(cell_image_files)
    print(cell_image_files)
    num_cell_images = len(cell_image_files)
    data = np.zeros((num_cell_images,50,50,3))
    count = 1
    for cell_image_file in cell_image_files:
        cell_image = io.imread(path.join(cell_images_path, cell_image_file))
        dst = cv2.fastNlMeansDenoisingColored(cell_image,None,10,10,7,29)
        resized_image = cv2.resize(dst, (256, 256))
        ycrcb=cv2.cvtColor(resized_image,cv2.COLOR_BGR2YCR_CB)
        channels=cv2.split(ycrcb)
        cv2.equalizeHist(channels[0],channels[0])
        cv2.merge(channels,ycrcb)
        histogram_image = cv2.cvtColor(ycrcb,cv2.COLOR_YCR_CB2BGR,resized_image)
        cv2.imwrite(fr'C:\Users\Admin\Desktop\Statistical Moments using Bins\Histogram Transformed Images\covid\transformed_image_covid_{count}.png',histogram_image)
        count +=1
    data = data.astype('float32')
    return data
```
