```python
df3 = bins.copy()
```

```python
import numpy as np
df3['log_Bins0'] = np.log10(df3['Bins0'])
df3['log_Bins1'] = np.log10(df3['Bins1'])
df3['log_Bins2'] = np.log10(df3['Bins2'])
df3['log_Bins3'] = np.log10(df3['Bins3'])
df3['log_Bins4'] = np.log10(df3['Bins4'])
df3['log_Bins5'] = np.log10(df3['Bins5'])
df3['log_Bins6'] = np.log10(df3['Bins6'])
df3['log_Bins7'] = np.log10(df3['Bins7'])

df3 = df3.drop(columns = ['Bins0', 'Bins1','Bins2','Bins3','Bins4', 'Bins5','Bins6', 'Bins7'])
```

```python
df3
```

```python
df3 = df3.drop(columns = ['Class'])
```

```python
df3['Class'] = bins['Class']
df3
```

```python
sns.pairplot(df3, hue="Class")
```

```python
sns.jointplot("log_Bins0", "log_Bins7", data=df3, kind="reg")
```

```python
corr = bins.corr(method = 'pearson')
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(10, 275)

sns.heatmap(corr, cmap=cmap, square=True, linewidths=0.5, cbar_kws={"shrink": 0.5}, ax=ax, annot=True)
```

```python
corr = df3.corr(method = 'pearson')
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(10, 275)

sns.heatmap(corr, cmap=cmap, square=True, linewidths=0.5, cbar_kws={"shrink": 0.5}, ax=ax, annot=True)
```

```python
from scipy.stats import pearsonr
x = df3['log_Bins0'][:]
y = df3['log_Bins7'][:]
cor1,_ = pearsonr(x, y)
print(f'Pearson co-efficient of Bins 0 and Bins 7: {cor1}')
```
