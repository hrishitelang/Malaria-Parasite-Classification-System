```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bins = pd.read_csv("image_bins_all.csv")
bins_p = pd.read_csv("image_bins_p_final.csv")
bins_np = pd.read_csv("image_bins_np_final.csv")
```

```python
bins_p.drop('filename', axis=1, inplace=True)
bins_np.drop('filename', axis=1, inplace=True)
```

```python
corr = bins_p.corr(method = 'pearson')
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(10, 275)

sns.heatmap(corr, cmap=cmap, square=True, linewidths=0.5, cbar_kws={"shrink": 0.5}, ax=ax)
```

```python
corr = bins_np.corr(method = 'pearson')
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(10, 275)

sns.heatmap(corr, cmap=cmap, square=True, linewidths=0.5, cbar_kws={"shrink": 0.5}, ax=ax)
```
